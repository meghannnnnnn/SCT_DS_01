```
import pandas as pd
import matplotlib.pyplot as plt
```

```
path="/content/drive/MyDrive/API_SP.POP.TOTL_DS2_en_excel_v2_31647.xls"
df=pd.read_excel(path,skiprows=3)
```

df = pd.read_excel(path, skiprows=3) reads an Excel file located at path and creates a pandas DataFrame named df.The skiprows=3 argument instructs pandas to skip the first 3 rows of the Excel file before starting to read the data.

```
df.columns = df.columns.str.strip()
```

```
country_name='India'
df_country = df.loc[df['Country Name'] == country_name]
```

```
df_country = df_country.iloc[:, 4:].transpose()
```

```
df_country.columns = ['Population Growth']
df_country.index.name = 'Years'
```

```
df_country.index = df_country.index.astype(int)
```

```
plt.figure(figsize=(10,10))
colors=['blue','red','yellow','green','orange','brown','grey','skyblue','purple','magenta','olive','indigo','lime','maroon','navy','lime','silver','gold']
plt.bar(df_country.index, df_country['Population Growth'], color=colors)
plt.title(f'Population Growth Over Time in {country_name}')
plt.xlabel('Year')
plt.ylabel('Population Growth')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', linewidth=0.10)
plt.tight_layout()
plt.show()
```
